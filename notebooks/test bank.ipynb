```python
firstnames = ('Hans','Peter','Maria','Sofie','Lars')
```

```python
lastnames = ('Müller','Meier','Hanssen','Oberberger')
```

```python
accounts = []
for firstname in firstnames:
    for lastname in lastnames:
        accountx = {'number':len(Bank1.accounts)+len(accounts), 'firstname': firstname, 'lastname': lastname}
        accounts.append(accountx)
```

```python
accounts
```

```python
for item in accounts:
    Bank1.open_account(item)
```
